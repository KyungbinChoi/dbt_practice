```python
# PostgreSQL 연결 설정
from sqlalchemy import create_engine
import pandas as pd
```

```python
# PostgreSQL 연결 URI
db_uri = "postgresql://airflow:airflow@postgres:5432/airflow"

# SQLAlchemy 엔진 생성
engine = create_engine(db_uri)

# SQL 쿼리 실행
query = "SELECT * FROM public_marts.monthly_sales;"
df = pd.read_sql(query, engine)
```
